```python
from spellchecker import SpellChecker
```

```python
def non_eng_mapper(text):
    words = re.findall(r'\w+', text.lower())  # converting to lowercase and extracting the code
    map_output = defaultdict(int)
    for word in words:
        map_output[word] += 1
    return map_output
```

```python
def non_eng_reducer(map_output):
    word_counts = defaultdict(int)
    for counts in map_output:
        for word, count in counts.items():
            word_counts[word] += count
    return word_counts
```

```python
def non_eng_words(word_counts, spell_checker):
    non_eng_cnt = defaultdict(int)
    for word, count in word_counts.items():
        if not spell_checker.known([word]):  # chencking for non_eng words
            non_eng_cnt[word] += count
    return non_eng_cnt
```

```python
# using file2
with open('file2.txt', 'r') as file2:
    text = file2.read()
```

```python
spell_check = SpellChecker() # initializing spell checker
```

```python
map_output = non_eng_mapper(text)
```

```python
word_counts = non_eng_reducer([map_output])
```

```python
non_eng_counts = non_eng_words(word_counts, spell_check) # getting the counts of non-english words
```

```python
for word, count in sorted(non_eng_counts.items(), key=lambda x: x[1], reverse=True):
    print(f'{word}: {count}') # displaying output
```
